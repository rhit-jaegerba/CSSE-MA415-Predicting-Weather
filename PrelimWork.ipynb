# Prelim Work on Weather Classification

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('weather_classification_data.csv')

df.shape
```

### Basic Exploration

```python
df.info()
```

There are 13200 entries in the dataset with 11 columns. There is a mixture of numerical and categorical features.

```python
# Display the first 5 rows of the dataset
df.head()
```

```python
# Check if target variable is balanced and what the possible values are.
print(df['Weather Type'].value_counts())
```

```python
df['Weather Type'].value_counts().plot(kind='bar')
plt.title('Distribution of Weather Types')
plt.xlabel('Weather Type')
plt.ylabel('Count')
plt.show()
```

We will be solving a classification problem on the Weather Type.  Weather Type is balanced with 4 catagories Rainy, Cloudy, Sunny, and Snowy, all with 3300 entries each.

```python
# Check for missing values in the dataset
print(df.isnull().sum())
```

The dataset is already balanced and cleaned of any null values. This dataset may already have been cleaned.

```python
print(df.describe())
```

Note that this dataset is synthetic and some outliers are beyond realistic measures. For example, some entries record humidity levels and precipitation levels above 100%, and temperatures (which are recorded in Celcius) record heats of upto 109 degrees (228 Fahrenheit).

```python
#Boxplots of all numeric features in the dataset
# List of columns to plot
columns_to_plot = ['Temperature', 'Humidity', 'Wind Speed', 'Precipitation (%)', 
                   'Atmospheric Pressure', 'UV Index', 'Visibility (km)']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 8))  # Adjust rows/cols as needed
axes = axes.flatten()

# Plot each boxplot
for i, column in enumerate(columns_to_plot):
    if i < len(axes):  # Ensure we don't exceed the number of axes
        df.boxplot(column=[column], ax=axes[i])
        axes[i].set_title(column)

# Hide any unused subplots
for j in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Calculate the mode (most frequent class) of the target variable
mode_class = df['Weather Type'].mode()[0]

# Predict the mode for all entries
baseline_predictions = [mode_class] * len(df)

# Calculate the baseline accuracy
from sklearn.metrics import accuracy_score

baseline_accuracy = accuracy_score(df['Weather Type'], baseline_predictions)
baseline_error_rate = 1 - baseline_accuracy

print(f"Baseline Accuracy: {baseline_accuracy:.2f}")
print(f"Baseline Error Rate: {baseline_error_rate:.2f}")
```

The baseline accuracy of 0.25 makes sense because we have an equal number of entries for each of our 4 weather types in our dataset.

The baseline accuracy sets a standard for our classification models. If our model accuracy is below 0.25, it is not a good model and is worse than simply predicting the mean.

### Correlation and Additional Plots

```python
from pandas.plotting import scatter_matrix

# Create scatter matrix
numeric_cols = df.select_dtypes(include=['number'])
scatter_matrix(numeric_cols, figsize=(12, 12), diagonal='hist', alpha=0.5)

plt.tight_layout()
plt.show()
```

There are a few things that can be inferred from these plots:
* UV index is discrete. Note the "lines" that mark the various UV index levels.
* The "blockiness" of the plots further exemplify the "fakeness" of the data: 
1. Take a look at the visibility vs precipitation graph. There is a block of data points with visibility 5~10 km when precipitation is below 50%, and there is another block of data points with visibility 0~5 km when precipication is above 50%.
2. Likewise, several plots have "step-function-like" behavior, i.e. humidity vs precipitation.
3. Take a look at the atmospheric pressure, where outliers are evenly distributed along the 800-1200 range instead of displaying normality.

```python
# Compute correlation only on numeric columns
corr_matrix = df.corr(numeric_only=True)

fig, ax = plt.subplots(figsize=(6, 5))
cax = ax.matshow(corr_matrix, cmap='coolwarm')

# Add colorbar and ticks
plt.colorbar(cax)
ax.set_xticks(range(len(corr_matrix.columns)))
ax.set_yticks(range(len(corr_matrix.columns)))
ax.set_xticklabels(corr_matrix.columns, rotation=90, fontsize=7)
ax.set_yticklabels(corr_matrix.columns, fontsize=7)

plt.title('Correlation Matrix', pad=10, fontsize=10)
plt.tight_layout(pad=1.0)
plt.show()
```

The correlational factors make sense. For example, humidity is expected to be high when it is raining (precipitation), which gives it a positive correlation (red). Likewise, visibility is expected to be high when it is not, which gives it a negative correlation (blue).

```python
# Define numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Loop through and plot
for num_col in numerical_cols:
    plt.figure(figsize=(4, 2))
    df.boxplot(column=num_col, by="Weather Type")
    plt.title(f'{num_col} by "Weather Type"')
    plt.suptitle('')  # suppress default title
    plt.xlabel("Weather Type")
    plt.ylabel(num_col)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

These boxplots show the correlation between weather type and numerical features. For example, rainy and snowy weather provide much less visiblity than cloudy and sunny days.

```python
# Define categorical columns
categorical_features = df.select_dtypes(include=['object']).columns

for cat_col in categorical_features:
    if cat_col in df.columns:
        # Create a cross-tabulation of counts
        ct = pd.crosstab(df[cat_col], df['Weather Type'])

        # Plot as grouped bar chart
        ct.plot(kind='bar', figsize=(6, 4))
        plt.title(f'{cat_col} vs Weather Type')
        plt.xlabel(cat_col)
        plt.ylabel('Count')
        plt.xticks(rotation=30, ha='right')
        plt.tight_layout()
        plt.show()
```

These bar graphs show the relationship between the weather types and categorical features. For example, sunny days are expected to be clear of clouds.
