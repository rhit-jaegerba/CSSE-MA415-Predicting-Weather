#### Clean dataset

```python
df = df.drop(['EventId', 'EndTime(UTC)', 'TimeZone', 'AirportCode', 'City', 'County', 'State', 'ZipCode'],axis=1)
df.head()
```

```python
df['StartTime(UTC)'] = pd.to_datetime(df['StartTime(UTC)'])

# Parse time into year, day of the year, and time of day (in minutes)
df['year'] = df['StartTime(UTC)'].dt.year
df['day'] = df['StartTime(UTC)'].dt.dayofyear
df['time'] = (
    df['StartTime(UTC)'].dt.hour * 60 +
    df['StartTime(UTC)'].dt.minute +
    df['StartTime(UTC)'].dt.second / 60
)

df.head()
```

```python
df = df.drop(['StartTime(UTC)'], axis=1)
```

```python
# Check if target variable is balanced and what the possible values are.
print(df['Type'].value_counts())
```

```python
# Combine precipitation and rain into a single weather type
df['Type'] = df['Type'].replace({'Precipitation': 'Rain'})
print(df['Type'].value_counts())
```
